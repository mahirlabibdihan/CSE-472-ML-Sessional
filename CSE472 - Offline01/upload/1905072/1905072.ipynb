# Data cleaning

## If you find any missing values in the dataset ( nan values) replace those data with the column wise mean

```python
# Select only numeric columns
numeric_columns = dataframe.select_dtypes(include=['number'])

# Fill null values with the mean of the respective columns
dataframe[numeric_columns.columns] = numeric_columns.fillna(numeric_columns.mean())
```

## If you find any duplicates in the dataset, keep just one copy of the data

```python
dataframe[dataframe.duplicated(keep=False)]
```

## if any row in the target column (Attrition) is missing, you must drop that row

```python
dataframe.dropna(subset=['Attrition'], inplace=True)
```

# Creation of input and output features

```python
features = dataframe.drop('Attrition', axis=1)
target = dataframe['Attrition']
target
```

# Conversion of features into numeric values

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
target = encoder.fit_transform(target)
target
```

```python
categorical_columns = ['BusinessTravel', 'Department', 'EducationField', 'JobRole',  'MaritalStatus', 'Over18', 'OverTime', 'Gender']

for col in categorical_columns:
    features[col] = features[col].astype('category')
    
```

```python
features.dtypes
```

```python
numeric_features = features.select_dtypes(include=['number'])
categorical_features = features.select_dtypes(include=['category'])
numeric_features
```

```python
categorical_columns_1 = ['BusinessTravel', 'Department', 'EducationField', 'JobRole',  'MaritalStatus']
categorical_columns_2 = ['Over18', 'OverTime', 'Gender'] # These columns have only two categories

categorical_features = pd.get_dummies(categorical_features, columns=categorical_columns_1)

for column in categorical_columns_2:
    categorical_features[column] = encoder.fit_transform(categorical_features[column])

categorical_features.dtypes
```

# Scaling of the features

```python
def scale(features, Scaler):
    scaler = Scaler()
    features_scaled = scaler.fit_transform(features)
    return features_scaled
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
features_scaled = scale(numeric_features,MinMaxScaler)
features_scaled
```

# Correlation Analysis

```python
features_df = pd.DataFrame(features_scaled, columns=numeric_features.columns)
features_df = pd.concat([features_df, categorical_features], axis=1)
target_df = pd.DataFrame(target, columns=['Attrition'])
features_df
```

```python
target_series = target_df['Attrition']
correlations = features_df.corrwith(target_series)
correlations
```

```python
top_20_columns = correlations.abs().sort_values(ascending=False).index[0:20]
top_20_features_df = features_df[top_20_columns]
correlations.abs().sort_values(ascending=False)[0:20]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

full_df = pd.concat([features_df, target_df], axis=1)
correlation_matrix = top_20_features_df.corr()

sns.heatmap(correlation_matrix,  cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
correlation_matrix
```

## 1D Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np 


# Separate the data based on binary labels
class_0 = top_20_features_df.loc[target_df["Attrition"] == 0]
class_1 = top_20_features_df.loc[target_df["Attrition"] == 1]

# Create a 1D scatter plot for sepal_width with binary labels
column = 'YearsWithCurrManager'
plt.figure()
plt.plot(class_0[column], np.zeros_like(class_0[column]), 'o', label='Class 0')
plt.plot(class_1[column], np.zeros_like(class_1[column]), 'o', label='Class 1')

plt.legend()
plt.xlabel(column)
plt.title(f'1D Scatter Plot of {column} by Binary Classes')
plt.show()
```

# Validating the pipeline (Bonus Task)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(top_20_features_df.to_numpy(), target, test_size=0.2, random_state=42)

# Step 2: Initialize the Logistic Regression classifier
clf = LogisticRegression()
# Step 3: Train the classifier on the training data
clf.fit(X_train, y_train)
# Step 4: Make predictions on the test set
y_pred = clf.predict(X_test)
# Step 5: Evaluate the classifier's performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Logistic Regression classifier: {accuracy}")
```
